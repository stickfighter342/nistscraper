```python
from nistscraper import rk_eos as rk
import numpy as np
import matplotlib.pyplot as plt
```

```python
rk.eos(1, 273, 0.01)
```

```python
vguess = np.logspace(-3, 0, 101)
res = rk.eos(1, 273, vguess)
print(rk.eos(1, 273, 0.029))
```

```python
fig, ax = plt.subplots()

ax.plot(vguess, res)
ax.set_ylim([-100, 100])
```

```python
pguess = np.logspace(-3, 4, 101)
res = rk.eos(pguess, 273, 0.1)
#print(rk.eos(1, 273, 0.029))
```

```python
fig, ax = plt.subplots()

ax.plot(pguess, res)
#ax.set_ylim([-100, 100])
ax.set_xlabel('P (atm)')
ax.set_ylabel('Residual of RK EOS')
```
